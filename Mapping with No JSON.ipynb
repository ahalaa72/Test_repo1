```python
import pandas as pd
from datetime import datetime
```

```python
def convertDateFormat(Column,old_format):
    Column=Column.apply(lambda x: datetime.strftime(datetime.strptime(x, old_format),'%Y%m%d'))
    return Column
```

```python
df=pd.read_excel('Sathapana USD - source.xlsx')
```

```python
df.columns=list('ABCDEFGHIJKLMNOPQ')
```

```python
lastRecordIndex=df.index[~df['A'].isna()][-1]
```

```python
firstRecordIndex=df.index[df['A']=='2'][0]
```

```python
linesDf=df.iloc[firstRecordIndex:lastRecordIndex+1].reset_index()
```

```python
linesDf['updatedDate']=[x[:11] for x in linesDf['B']]
```

```python
linesDf['updatedDate']=convertDateFormat(linesDf['updatedDate'],'%d-%b-%Y')
```

```python
Output=pd.DataFrame()
```

```python
Output['StatementDate']=[x[:11] for x in linesDf['B']]
```

```python
Output['StatementDate']=convertDateFormat(Output['StatementDate'],'%d-%b-%Y')
```

```python
Output['StatementOpeningLedger']=None
```

```python
Output['StatementOpeningAvailable']=None 
```

```python
Output['ValueDate']=[x[:11] for x in linesDf['D']]
```

```python
Output['ValueDate']=convertDateFormat(Output['ValueDate'],'%d-%b-%Y')
```

```python
Output['BankTransactionTypeCode']=['FCHG' if x in ['COM','MSC'] else 'FMSC' for x in linesDf['E'] ]
```

```python
Output['DebitCreditIndicator']=['Debit' if float(x) >0 else 'Credit' for x in linesDf['L']]
```

```python
Output['LineAmount']=-linesDf['L']+linesDf['O']
```

```python
Output['BankReferenceNumber']=linesDf['J']
```

```python
Output['RelatedReference']=None
```

```python
Output['TextReferenceNumber']=None
```

```python
Output['Description']=linesDf['G']
```

```python
Output['CashManagementAccount']=df.at[7,'N']
```

```python
Output['Intraday']='FALSE'
```

```python
Output['Currency']=df.at[2,'M'][-3:]
```

```python
Output['StatementClosingLedger']=None 
Output['StatementClosingAvailable']=None 
```

```python
dateSet=set(linesDf['updatedDate'])
```

```python
for date in dateSet:
    Output.loc[Output.StatementDate==date,'StatementOpeningLedger']=float(df.at[linesDf.index[linesDf.updatedDate==date][0]+firstRecordIndex-1,'P'])
    Output.loc[Output.StatementDate==date,'StatementOpeningAvailable']=float(df.at[linesDf.index[linesDf.updatedDate==date][0]+firstRecordIndex-1,'P'])  
    Output.loc[Output.StatementDate==date,'StatementClosingLedger']=float(df.at[linesDf.index[linesDf.updatedDate==date][-1]+firstRecordIndex,'P'])
    Output.loc[Output.StatementDate==date,'StatementClosingAvailable']=float(df.at[linesDf.index[linesDf.updatedDate==date][-1]+firstRecordIndex,'P'])
```

```python
columns=['CashManagementAccount','Intraday','StatementDate','Currency','StatementOpeningLedger','StatementClosingLedger','StatementOpeningAvailable','StatementClosingAvailable','ValueDate','BankTransactionTypeCode','DebitCreditIndicator','LineAmount','BankReferenceNumber','RelatedReference','TextReferenceNumber','Description']
```

```python
Output=Output[columns]
```
